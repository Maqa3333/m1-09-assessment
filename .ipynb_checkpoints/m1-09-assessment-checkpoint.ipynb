```python





# TODO: clean pm25
df["pm25"] = pd.to_numeric(df["pm25"], errors="coerce")
after_na = df.isna().sum()
after_na
invalid_converted = after_na - before_na
invalid_converted
#I checked that the pm25 column is numeric value or not.If its type is string,then it returns error
# TODO: handle missing values in pm25 and justify in markdown
df_clean = df.dropna(subset = ["pm25"])
df_clean.shape
df_clean.tail()
```

## Part B: Required Analysis (TODOs)

- Data quality analysis: city with highest % invalid/missing pm25.
- Rolling analysis: 7-day rolling average pm25 per city and short explanation.
- Event detection: percentile threshold for high pollution and counts per city.
- Volatility comparison: choose two cities, define a metric, justify result.
- Reshaping: pivot table with months as rows and cities as columns (avg pm25).

```python
# TODO: analysis tasks
# - data quality by city
missing_percentage = (
    df["pm25"].isna().groupby(df["city"]).mean()*100
                     )
missing_percentage.sort_values(ascending = False)
# - rolling averages per city
rolling_average = (
    df.groupby("city")["pm25"].rolling(window = 7).mean().reset_index(level =0,drop=True)
                  )
print(rolling_average)
# - percentile-based event counts
threshold = df["pm25"].quantile(0.90)
high_pollution_events = df[df["pm25"] > threshold].groupby("city").size()
print(threshold)
print(high_pollution_events)
# - volatility comparison
volatility = df.groupby("city")["pm25"].std()
print(volatility)
# - pivot table by month and city
pivot_monthly = df.pivot_table(values="pm25", index=df.index.month, columns="city", aggfunc="mean")
pivot_monthly.index.name = "Month"
print(pivot_monthly)
```

## Part C: Aggregations (TODOs)

- Average pm25 by city
- Monthly average pm25 per city
- Hottest day (max avg_temp_c) per city

```python
# TODO: aggregation tasks
# avg_pm25_by_city = ...
avg_pm25_by_city = df.groupby("city")["pm25"].mean()
# monthly_avg_pm25 = ...
monthly_avg_pm25 = df.groupby(["city", pd.Grouper(freq="ME")])["pm25"].mean()
# hottest_day_per_city = ...
hottest_day_idx = df.groupby("city")["avg_temp_c"].idxmax()
hottest_day_per_city = df.loc[hottest_day_idx, ["city", "avg_temp_c"]]
```

```python
# I calculated the overall average PM2.5 level for each city.
# I grouped the data to find the monthly average pollution for each city.
# I identified the specific date with the maximum temperature for every city.
```

```python
# TODO: plotting section
plt.figure(figsize=(10, 5))
pivot_monthly.plot(kind="line", marker="o", ax=plt.gca())
plt.title("Monthly Average PM2.5 Trend by Cities")
plt.ylabel("Average PM2.5")
plt.xlabel("Months (1=January, 6=June)")
plt.legend(title="Cities")
plt.show()

# I plotted monthly trends to see how pollution changes throughout the year.
```

```python
plt.figure(figsize=(8, 4))
avg_pm25_by_city.sort_values(ascending=False).plot(kind="bar", color="coral", edgecolor="black")
plt.title("Overall Average PM2.5 for Cities")
plt.ylabel("PM2.5 Level")
plt.xticks(rotation=0)
plt.show()
# I created a bar chart to rank cities by their overall average pollution.
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["avg_temp_c"], df["pm25"], alpha=0.4, c="blue")
plt.title("Relationship between Temperature and Air Pollution (PM2.5)")
plt.xlabel("Average Temperature (°C)")
plt.ylabel("PM2.5 Level")
plt.show()
# I made a scatter plot to check if higher temperatures lead to lower pollution.
```
